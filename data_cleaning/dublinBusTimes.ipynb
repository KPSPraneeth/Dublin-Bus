```python
busTimes = {}
```

```python
for a in soup.find_all('a', href=True):
    if a['href'][0:41] == "/Your-Journey1/Timetables/All-Timetables/":
        link = "https://www.dublinbus.ie/"+a['href']
        
        pageTest = requests.get(link).text
        soup2 = BeautifulSoup(pageTest, "lxml")
        
        totalTime = 0
        for a in soup2.find_all("span", style="color:#01aef0"):
            try:
                lineContents = str(a).split("»")
                timeText = lineContents[1]
                time = int(timeText.replace("mins","").strip())
                totalTime += time
            except:
                try:
                    lineContents = str(a).split("&gt")
                    timeText = lineContents[2]
                    time = timeText.replace("mins","").strip()
                    time = int(time.replace(";","").strip())
                    totalTime += time
                    
                except:
                    pass
        
        idBus = soup2.find_all("title")[0].text.strip()
        
        #print(link)
        print(idBus[0:3] , totalTime/2)
        busTimes[idBus[0:3]] = totalTime/2
    
```

```python
busTimes
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(busTimes, orient='index')
```

```python
df.rename(columns={0:"timeTotal"}, inplace=True)
```

```python
df
```

```python
df.to_csv("totalTimeRoutes.txt")
```
